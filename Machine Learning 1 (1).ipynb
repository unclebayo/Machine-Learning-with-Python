## Linear Regression
## Simple LR = Predicting Price using number of rooms
i.e. predicting price with just one variable
This Simple Linear Regression is to predict Price(Y) using one independent variable room_num as X

```python
#First import the library to do the linear regression - statsmodel.api
import statsmodels.api as sn
```

```python
#add constants x and B0 as seen in Simple Linear Regression Formula
X = sn.add_constant(df['room_num'])
```

```python
#to fit the model, we create a variable using OLS for the dependent variable
lm = sn.OLS(df['price'], X).fit()
```

```python
#get summary of the model
lm.summary()
```

1. There is a significant relationship between room_num and price because the p values [P>|t|] are close to zero
2. B1 = co-eff of room number = 9.0997
3. B0 = co-eff of const = -34.6592 = intercept
4. R squared is okay because our value is close to 0.5

## Multiple Regression
## Predict Price using all variables

```python
#first method
import statsmodels.api as sn
```

```python
#since we are using all independent variables to predict price, we need to define x and drop price
#we want all variables except the price variable
#since we want to drop a column, we use axis = 1
#axis = 0 is for dropping rows, axis= 1 is for dropping columns
x_multi = df.drop("price", axis =1)
```

```python
#to check if the dependent variable, y=price is dropped
x_multi.head()
```

```python
#to define y
y_multi = df['price']
```

```python
y_multi.head()
```

```python
#add a constant to x
x_multi_const = sn.add_constant(x_multi)
```

```python
x_multi_const.head()
```

```python
#fit our model
lm_multi = sn.OLS(y_multi, x_multi_const).fit()
```

```python
lm_multi.summary()
```

1. The R-Squared value is greater than 0.5 meaning that the variables are statistically okay to predict the price
2. Degree of Freedom is 490. Gotten by n-t-1. Where n = 506 = number of obs., t= number of columns = 15 = no. of variables i.e. 506-15-1

# Test and Train Split
## Splitting the dataset into Test and Train

```python
#import the f
from sklearn.model_selection import train_test_split
```

```python
#define your four variables
#dependent and independent train variables (y and x)
#dependent and independent test variables (y and x)
#the ratio for division inro test and train is 80:20
x_train, x_test, y_train, y_test = train_test_split(x_multi, y_multi, test_size = 0.2, random_state = 0)
```

```python
#to get the number of rows and columns in all variables
print (x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#creating linear regression
lm_a = LinearRegression()
```

```python
#fit the model
lm_a.fit(x_train, y_train)
```

```python
#predict values of of the test set
y_test_a = lm_a.predict(x_test)
```

```python
#predict values of of the test set
y_train_a = lm_a.predict(x_train)
```

```python
#to check the R-squared value for our training and test data, we will import
from sklearn.metrics import  r2_score
```
